```python
import base64

class HYPER_Grader:
    def __init__(self):
        self.answers = {
            1: "WzMsIDQsIDVd",        # [3, 4, 5]
            2: "MjU=",                # 25
            3: "Sm9obg==",            # John
            4: "Mw==",                # 3
            5: "VHJ1ZQ=="             # True
        }

    def check(self, task_id, user_answer):
        try:
            expected = base64.b64decode(self.answers[task_id]).decode()
            result = str(user_answer).strip()
            expected = expected.strip()

            if result == expected:
                print(f"Task {task_id}: ✅ 정답입니다!")
            else:
                print(f"Task {task_id}: ❌ 틀렸습니다.")
                print("당신의 출력:", result)
                print("정답:", expected)

        except Exception as e:
            print(f"Task {task_id}: ⚠ 에러 발생 ({e})")

grader = HYPER_Grader()
print("채점 준비 완료!")
```
