```python
import base64

class HYPER_Grader:
    def __init__(self):
        self.answers = {
            1: "NA==",   
            2: "Mw==",   
            3: "aGVsbG9fd29ybGQ=",  
            4: "UG4=",   
            5: "Mw=="    
        }

    def check(self, task_id, user_func):
        try:
            expected = base64.b64decode(self.answers[task_id]).decode()
            result = str(user_func()).strip()
            expected = expected.strip()

            if result == expected:
                print(f"Task {task_id}: ✅ 정답입니다!")
            else:
                print(f"Task {task_id}: ❌ 틀렸습니다.")
                print("당신의 출력:", result)
                print("정답:", expected)

        except Exception as e:
            print(f"Task {task_id}: ⚠ 에러 발생 ({e})")

grader = HYPER_Grader()
print("채점 준비 완료!")
```
