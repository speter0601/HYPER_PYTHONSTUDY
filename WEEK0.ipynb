```python
# 자동 채점기
import base64

class KHUDA_Grader:
    def __init__(self):
        self.answers = {
            1: "SGVsbG8gUHl0aG9uIQ==",  # Hello Python!
            2: "OA==",                 # 8
            3: "Ng=="                  # 6
        }

    def check(self, task_id, user_func):
        try:
            expected = base64.b64decode(self.answers[task_id]).decode()
            result = str(user_func()).strip()
            expected = expected.strip()

            if result == expected:
                print(f"Task {task_id}: ✅ 정답입니다!")
            else:
                print(f"Task {task_id}: ❌ 틀렸습니다.")
                print("당신의 출력:", result)
                print("정답:", expected)

        except Exception as e:
            print(f"Task {task_id}: ⚠ 에러 발생 ({e})")

grader = KHUDA_Grader()
print("채점 준비 완료!")
```

## Task 1 — 문자열 출력

Hello Python! 을 출력하는 프로그램을 작성하세요.

입력 없음  
출력:  
Hello Python!

```python
def solution1():
    result = "Hello Python!"

    return result

grader.check(1, solution1)
```

## Task 2 — 두 숫자의 합

두 정수를 입력받아 합을 출력하세요.

입력 예시:
3 5

출력:
8

힌트: input().split()

```python
def solution2():
    a, b = map(int, input().split())
    result = a + b


    return result

grader.check(2, solution2)
```

## Task 3 — 문자열 길이

문자열을 입력받아 길이를 출력하세요.

입력 예시:
python

출력:
6

```python
def solution3():
    text = input()
    result = len(text)


    return result

grader.check(3, solution3)
```
